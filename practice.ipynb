```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://news.ltn.com.tw/list/BreakingNews')
soup = BeautifulSoup(res.text)
for ele in soup.select('.lipic'):
    print ele.select('.picword')[0].text, ele.select('span')[0].text
```

```python
import requests
from bs4 import BeautifulSoup as bs
for pg in range(0,1):
    res = requests.get('http://www.ipeen.com.tw/search/taoyuan/000/1-0-2-0/?p={0}'.format(pg))
    soup = bs(res.text)
    for ele in soup.select('.serShop'):
        print " 店名:",ele.select('.name a')[0].text, " 地址:",ele.select('span')[1].text.strip()
```
